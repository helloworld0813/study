```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
#형태소를 분석해 주는 모듈
from konlpy.tag import Twitter
twitter = Twitter()
```

```python
#원댓
barcode = input('barcode를 입력하세요:')

def get_reple(pageNumber=1):
    
    url = f'http://www.kyobobook.co.kr/product/productSimpleReviewSort.laf?gb=klover&ejkGb=KOR&mallGb=&sortType=like&orderType=order&barcode={barcode}&pageNumber={pageNumber}'

    resp = requests.get(url)
    soup = BeautifulSoup(resp.text,"lxml")
    page=soup.select_one('ul.board_list').select('div.comment_wrap')
    
    r, s =[],[]
    
    for dl in page:
        #print('평점:',dl.span.text)
        #print('댓글:', dl.select('div.txt'))
        reviews=dl.select('div.txt')  
        for review in reviews:
            #print('댓글:', review.text)
            review=review.text
            review = review.replace('\n','')
            review = review.replace('\r','')
            if int(dl.span.text) >=4:
                r.append(review)
                s.append(1) # 긍정
            elif int(dl.span.text) <=2:
                r.append(review)
                s.append(0) #부정
    return r,s
                
```

```python
#전체페이지 개수 알아내기
maximum = 0
page = 1
while 1:
    url = f'http://www.kyobobook.co.kr/product/productSimpleReviewSort.laf?gb=klover&ejkGb=KOR&mallGb=&sortType=like&orderType=order&barcode=9791158510244&pageNumber=1'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text,"lxml")
    page_list = soup.findAll("div", {"class": "paging_num" +str(page)})
    
    if not page_list:
        maximum = page - 1
        break
    page = page + 1
print("총 " + str(maximum) + " 개의 페이지가 확인 됬습니다.")
```

```python
#원댓용
import time
reple, score = [], []
#여기서 i는 페이지 수
for i in range(1,130):
    time.sleep(1)
    print('요청횟수:',i, end='\r')
    get_reple(i)
    r, s = get_reple(i)
    reple += r    
    score += s
```

##트레이닝/ 테스트 셋 분리하기

```python
from sklearn.model_selection import train_test_split #자동완성기능
```

```python
#8대2로 나눠줌
train_x, test_x, train_y, test_y = train_test_split(reple, score, test_size=0.2, random_state=0) 
```

```python
len(train_x), len(train_y)
```

```python
len(test_x), len(test_y)
```

```python
train_x
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#ngeam => 1그램, 2그램 단어를 1단어에서 2단어로 나워서 보겠다. 이런의미
#min_df => 최소 3개나오는 것을 기준으로
#max_df=> 너무 많은 빈도수 는 특징이 될 수 없다고 판단하여 0.9를 기준치로 잡음 

tfv = TfidfVectorizer(tokenizer=twitter.morphs, ngram_range=(1,2), min_df=3, max_df=0.9) #빈도수 너무 높은것 제외, 너무 작은것도 제외
tfv.fit(train_x)
tfv_train_x=tfv.transform(train_x) #학습한것 적용하기
tfv_train_x
```

```python
from sklearn.linear_model import LogisticRegression#이진 분류 알고리즘
from sklearn.model_selection import GridSearchCV #하이퍼마라미터
```

```python
#cv=> cross valudation으로 자기 스스로 평가하도록 하는것
clf=LogisticRegression(random_state=0)
params =  {'C':[1,3,5,7,9]}
#좋은 걸 서치해서 알려주기, 정확도가 높은것으로
grid_cv=GridSearchCV(clf, param_grid=params, cv=4, scoring='accuracy', verbose=1)
grid_cv.fit(tfv_train_x, train_y)
```

```python
#적합하 것 확인하기 => 1번으로 확인
grid_cv.best_params_ 
```

```python
#10번 중에 9번은 맞춘다는 의미
grid_cv.best_score_
```

```python
# 테스트하기, 정확도 측정
tfv_test_x  = tfv.transform(test_x)
grid_cv.best_estimator_.score(tfv_test_x, test_y)
```

```python
# 1이 나오면 긍정 => 부정 분류가 안되네
a=['아 너무 재밌어요. 꼭 보세요',
  '핵노잼 너무 재미없어, 절대 보지마.',
  '영화 너무 훌륭해요',
  '이걸 영화라고',
  '영화가지루해요',
  '보지마세요',
  '이걸 책이라고 만드냐',
  '이 책 보지마세요 진짜 재미없어요']
my_review=tfv.transform(a)
grid_cv.best_estimator_.predict(my_review)
```
